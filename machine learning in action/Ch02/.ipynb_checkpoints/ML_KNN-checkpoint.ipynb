# 概述

K-近邻算法采用测量不同特征值之间的距离方法进行分类  
  
工作原理：  
存在一个每个数据都存有标签的样本数据集合，输入没有标签的新数据后，将新数据的每个特征与样本集中数据对应的特征进行比较，然后算法提取样本集中特征最相似数据（最近邻）的分类标签。一般来说只选择样本数据集中前k个最相似的数据，通常k是不大于20的整数。最后选择k个最相似数据中出现次数最多的分类，作为新数据的分类。  
  
优先：精度高、对异常值不敏感、无数据输入假定  
缺点：计算复杂度高、空间复杂度高  
适用数据范围：数值型和标称型

# 代码实现

## 导入数据

```python
import numpy as np
import operator # 运算符模块
```

```python
def createDataSet():
    group = np.array([[1.0, 1.1], [1.0, 1.0], [0, 0], [0, 0.1]])
    labels = ['A', 'A', 'B', 'B']
    return group, labels
```

## 构造分类器

```python
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    diffMat = np.tile(inX, (dataSetSize, 1)) - dataSet # inX沿Y轴复制dataSetSize倍
    sqDiffMat = diffMat ** 2 # 平方误差
    sqDistances = sqDiffMat.sum(axis=1) #按行相加，结果为dataSetSize大小的数组
    distances = sqDistances ** 0.5
    sortedDistIndicies = distances.argsort() # 返回从小到大排序后的索引值
    classCount = {}
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        # 对标签进行统计，dict.get()进行查找对应标签的值，如果不存在就为0，最后加1
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
    # operator.itemgetter 按照第二元素的次序对元组进行排序
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1),
                              reverse=True)
    return sortedClassCount[0][0]
```

```python
classify0([0, 0], group, labels, 3)
```

## 导入数据

```python
# filename = 'datingTestSet.txt'
def file2matrix(filename):
    with open(filename) as f:
        data = f.readlines()
    numberOfLines = len(data)
    dataMat = np.zeros((numberOfLines, 3)) # 参数(x, y)为生成一个x行y列的零矩阵
    classLabelVector = []
    index = 0
    for line in data:
        line = line.strip() # 移除首位的换行符
        listFromLine = line.split('\t')
        dataMat[index, :] = listFromLine[0:3] # 存储特征
        classLabelVector.append(int(listFromLine[-1])) # 存储标签
        index += 1
    return dataMat, classLabelVector
```

```python
datingDataMat, datingLabels = file2matrix('datingTestSet2.txt')
```

## 归一化

newVale = (oldValue - min) / (max - min)

```python
def autoNorm(dataSet):
    minVals = dataSet.min(0) # 取每个特征的最小值，所以需要按列求最小
    maxVals = dataSet.max(0) # 结果为长度3的array
    valRange = maxVals - minVals
    normDataSet = np.zeros(np.shape(dataSet))
    m = dataSet.shape[0] # 样本数量
    normDataSet = dataSet - np.tile(minVals, (m, 1))
    normDataSet = normDataSet / np.tile(valRange, (m, 1))
    return normDataSet, valRange, minVals
```

```python
normMat, ranges, minVals = autoNorm(datingDataMat)
```

## 算法测试

```python
def datingClassTest(is_print=False):
    hoRatio = 0.1 # 测试集比例
    datingDataMat, datingLabels = file2matrix('datingTestSet2.txt') # 读取数据
    normMat, ranges, minVals = autoNorm(datingDataMat) # 特征归一化
    m = normMat.shape[0] # 样本总量
    numTestVecs = int(m*hoRatio) # 测试集数量
    errorCount = 0
    for i in range(numTestVecs):
        classifierResult = classify0(normMat[i, :], normMat[numTestVecs:m, :],
                                     datingLabels[numTestVecs:m], 3)
        if is_print:
            print("the classifier came back with: %d, the real answer is %d" \
                  % (classifierResult, datingLabels[i]))
        if classifierResult!=datingLabels[i]:
            errorCount += 1
    print("the total error rate is: %.4f" % (errorCount/float(numTestVecs)))
```

```python
datingClassTest()
```

## 实例预测

```python
def classifyPerson(percentTats, ffMiles, iceCream):
    resultList = ['not at all', 'in small doses', 'in large doses']
    datingDataMat, datingLabels = file2matrix('datingTestSet2.txt') # 读取数据
    normMat, ranges, minVals = autoNorm(datingDataMat) # 特征归一化
    inArr = np.array([percentTats, ffMiles, iceCream])
    classifierResult = classify0((inArr-minVals)/ranges, normMat,datingLabels, 3)
    print("You will probably like this person: ", resultList[classifierResult-1])
```

```python
classifyPerson(10, 10000, 0.5)
```
